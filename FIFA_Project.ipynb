```python
player_salary = fifa[["short_name","wage_eur"]]
```

```python
player_salary.head()
```

# Top 10 players with higher salary

```python
player_salary = player_salary.sort_values(by=["wage_eur"],ascending = False)
```

```python
player_salary.head(10)
```

# The Plot of top 5 players with higher salary

```python
plt.figure(figsize = (10,5))
plt.bar(list(player_salary["short_name"])[0:5],list(player_salary["wage_eur"])[0:5], color=["green","blue","pink","orange","red"])
plt.show()
```
